```python
import requests
import pandas as pd
import textstat
import spacy
import os
from google.colab import files, drive
from sentence_transformers import SentenceTransformer
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Mount Google Drive
drive.mount('/content/drive')

# Load NLP models for semantic similarity
nlp = spacy.load("en_core_web_md")
sbert_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Define API details
API_URL = "https://openrouter.ai/api/v1/chat/completions"
API_KEY = "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" # Load API key from environment variables

def rewrite_content(text, model="openai/gpt-4-turbo", reference_examples=[]):
    """
    Uses OpenRouter AI to rewrite content while maintaining the original article's tone and ensuring a human-like output.
    """
    # Format reference examples for AI fine-tuning
    example_texts = "\n\n".join([f"Original: {ex['original']}\nRewritten: {ex['rewritten']}" for ex in reference_examples])

    prompt = f"""
    ### **üöÄ Your Role:**
    You are a professional business writer. Your task is to rewrite the given text to be clear, professional, and engaging.

    ---
    ### **üü¢ Example Rewrites from Trained Dataset:**
    {example_texts}

    ---
    ### **üìù Original Text:**
    "{text}"

    ---
    ### **üöÄ Your Task:**
    Rewrite the text **while matching the tone of the dataset examples**.
    Ensure the content is **human-like, natural, and easy to read**.
    - Rewrite this text in **simple, clear, and professional language**.
    - Avoid **business jargon and overly complex phrasing**.
    - Keep sentences **short and direct** (avoid long, multi-clause sentences).
    - Ensure readability score is **above 50**.
    - Maintain the **original meaning and tone**.
    """

    headers = {"Authorization": f"Bearer {API_KEY}", "Content-Type": "application/json"}
    payload = {"model": model, "messages": [{"role": "user", "content": prompt}]}

    response = requests.post(API_URL, json=payload, headers=headers)

    if response.status_code == 200:
        return response.json()["choices"][0]["message"]["content"]
    else:
        return f"API Request Failed: {response.json()}"

def validate_rewritten_content(original, rewritten, reference_dataset):
    """
    Validates rewritten content by checking:
    1. Meaning retention using BERT-based similarity.
    2. Tone adaptation using cosine similarity with reference dataset.
    3. Readability score to ensure content is clear and accessible.
    """
    # Compute BERT-based semantic similarity
    original_embedding = sbert_model.encode(original, convert_to_tensor=True)
    rewritten_embedding = sbert_model.encode(rewritten, convert_to_tensor=True)
    similarity_score = float(original_embedding @ rewritten_embedding.T)

    # Compute tone similarity using dataset examples
    dataset_embeddings = sbert_model.encode(reference_dataset, convert_to_tensor=True)
    tone_similarity_scores = [float(rewritten_embedding @ ref.T) for ref in dataset_embeddings]
    avg_tone_similarity = sum(tone_similarity_scores) / len(tone_similarity_scores)

    # Readability scores
    original_score = textstat.flesch_reading_ease(original)
    rewritten_score = textstat.flesch_reading_ease(rewritten)

    return {
        "Semantic Similarity": f"{similarity_score:.2f}",
        "Tone Similarity to Dataset": f"{avg_tone_similarity:.2f}",
        "Meaning Retained": similarity_score > 0.80,
        "Tone Matched": avg_tone_similarity > 0.75
    }

# Test the rewriting function
models = ["openai/gpt-4"]
original_text = """How AI Can Help in Sales to Businesses

In today‚Äôs fast-paced business world, staying ahead in sales can be challenging. This is where Artificial Intelligence (AI) steps in as a game-changer. AI might sound like something out of a sci-fi movie, but it‚Äôs simply a set of tools that can help businesses work smarter, not harder.

First, let‚Äôs talk about finding the right customers. AI can analyze data from various sources to identify people or businesses most likely to buy your product. Think of it as having a crystal ball for your sales team! Instead of chasing every lead, you focus on the ones that matter, saving time and effort.

Next, AI can improve the way you talk to customers. Tools powered by AI can study customer behavior and suggest the best time to call or email them. Imagine knowing when a potential customer is most likely to pick up the phone or open your email‚Äîit‚Äôs like having a sales coach guiding you.

AI can also make sales teams more productive. It can take care of repetitive tasks like updating customer information, scheduling follow-ups, or even writing personalized messages. This means your team has more time to build relationships and close deals.

Finally, AI helps businesses make smarter decisions. By analyzing past sales trends, AI can predict what products will sell more in the future or which areas need more focus. This helps you plan better and avoid unnecessary risks.

In short, AI is not about replacing sales teams‚Äîit‚Äôs about empowering them to perform better. For businesses looking to grow, using AI in sales is like having an extra set of hands and eyes working tirelessly in the background. It‚Äôs the competitive edge you need to succeed!
"""

# Sample reference dataset for tone verification
reference_dataset = [
    "AI is transforming sales by helping businesses focus on valuable leads, automate tasks, and improve decision-making.",
    "With AI, businesses can predict sales trends, streamline operations, and enhance customer engagement.",
    "Leveraging AI in sales helps teams work smarter, not harder, making the entire process more efficient."
]

for model in models:
    rewritten_text = rewrite_content(original_text, model)
    validation_result = validate_rewritten_content(original_text, rewritten_text, reference_dataset)

    print("\n=== Original Article ===\n", original_text)
    print("\n=== Rewritten Content ===\n", rewritten_text)
    print("\n=== Validation Results ===\n", validation_result)
```

```python
import requests
import pandas as pd
import spacy
import os
from google.colab import files, drive
from sentence_transformers import SentenceTransformer
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Mount Google Drive
drive.mount('/content/drive')

# Load NLP models
nlp = spacy.load("en_core_web_md")
sbert_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Load API key securely
API_URL = "https://openrouter.ai/api/v1/chat/completions"
API_KEY = "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"# ‚úÖ Secure API key loading

def rewrite_content(text, model="openai/gpt-4-turbo", reference_examples=[]):
    """
    Uses OpenRouter AI to rewrite content while maintaining the original article's tone and ensuring a human-like output.
    """
    # Format reference examples for AI fine-tuning
    example_texts = "\n\n".join([f"Original: {ex['original']}\nRewritten: {ex['rewritten']}" for ex in reference_examples])

    prompt = f"""
    ### **üöÄ Your Role:**
    You are a professional business writer. Your task is to rewrite the given text to be clear, professional, and engaging.

    ---
    ### **üü¢ Example Rewrites from Trained Dataset:**
    {example_texts}

    ---
    ### **üìù Original Text:**
    "{text}"

    ---
    ### **üöÄ Your Task:**
    Rewrite the text **while matching the tone of the dataset examples**.
    Ensure the content is **human-like, natural, and easy to read**.
    - Rewrite this text in **simple, clear, and professional language**.
    - Avoid **business jargon and overly complex phrasing**.
    - Maintain the **original meaning and tone**.
    - Maintain the original sentence structure as much as possible while improving clarity.
    - Avoid unnecessary synonyms that alter the core meaning.

    """

    headers = {"Authorization": f"Bearer {API_KEY}", "Content-Type": "application/json"}
    payload = {"model": model, "messages": [{"role": "user", "content": prompt}]}

    # ‚úÖ Error handling for API failures
    try:
        response = requests.post(API_URL, json=payload, headers=headers)
        response.raise_for_status()  # Raise an error for failed API calls
        return response.json()["choices"][0]["message"]["content"]
    except requests.exceptions.RequestException as e:
        return f"API Request Failed: {e}"

def validate_rewritten_content(original, rewritten, reference_dataset):
    """
    Validates rewritten content by checking:
    1. Meaning retention using BERT-based similarity.
    2. Tone adaptation using cosine similarity with reference dataset.
    """
    # ‚úÖ Encode dataset once (efficiency improvement)
    dataset_embeddings = sbert_model.encode(reference_dataset, convert_to_tensor=True)

    # Compute BERT-based semantic similarity
    original_embedding = sbert_model.encode(original, convert_to_tensor=True)
    rewritten_embedding = sbert_model.encode(rewritten, convert_to_tensor=True)
    similarity_score = float(original_embedding @ rewritten_embedding.T)

    # Compute tone similarity
    tone_similarity_scores = [float(rewritten_embedding @ ref.T) for ref in dataset_embeddings]
    avg_tone_similarity = sum(tone_similarity_scores) / len(tone_similarity_scores)

    return {
        "Semantic Similarity": f"{similarity_score:.2f}",
        "Tone Similarity to Dataset": f"{avg_tone_similarity:.2f}",
        "Meaning Retained": similarity_score > 0.80,
        "Tone Matched": avg_tone_similarity > 0.75
    }

# Test the rewriting function
models = ["openai/gpt-4"]
original_text = """How AI Can Help in Sales to Businesses

In today‚Äôs fast-paced business world, staying ahead in sales can be challenging. This is where Artificial Intelligence (AI) steps in as a game-changer. AI might sound like something out of a sci-fi movie, but it‚Äôs simply a set of tools that can help businesses work smarter, not harder.

First, let‚Äôs talk about finding the right customers. AI can analyze data from various sources to identify people or businesses most likely to buy your product. Think of it as having a crystal ball for your sales team! Instead of chasing every lead, you focus on the ones that matter, saving time and effort.

Next, AI can improve the way you talk to customers. Tools powered by AI can study customer behavior and suggest the best time to call or email them. Imagine knowing when a potential customer is most likely to pick up the phone or open your email‚Äîit‚Äôs like having a sales coach guiding you.

AI can also make sales teams more productive. It can take care of repetitive tasks like updating customer information, scheduling follow-ups, or even writing personalized messages. This means your team has more time to build relationships and close deals.

Finally, AI helps businesses make smarter decisions. By analyzing past sales trends, AI can predict what products will sell more in the future or which areas need more focus. This helps you plan better and avoid unnecessary risks.

In short, AI is not about replacing sales teams‚Äîit‚Äôs about empowering them to perform better. For businesses looking to grow, using AI in sales is like having an extra set of hands and eyes working tirelessly in the background. It‚Äôs the competitive edge you need to succeed!
"""

# Sample reference dataset for tone verification
reference_dataset = [
    "AI is transforming sales by helping businesses focus on valuable leads, automate tasks, and improve decision-making.",
    "With AI, businesses can predict sales trends, streamline operations, and enhance customer engagement.",
    "Leveraging AI in sales helps teams work smarter, not harder, making the entire process more efficient."
]

for model in models:
    rewritten_text = rewrite_content(original_text, model)
    validation_result = validate_rewritten_content(original_text, rewritten_text, reference_dataset)

    print("\n=== Original Article ===\n", original_text)
    print("\n=== Rewritten Content ===\n", rewritten_text)
    print("\n=== Validation Results ===\n", validation_result)
```
